# Procedual Programming Paradigm

####  This is the paradigm in which we imperatively instruct the program what to do using a procedure.
That is, firstly we write the steps of solving a particular problem, and then convert it into the code. This doesnot necessarily the same procedure as mathematical definition. 

### Characterstics
- Variables are used for denoting a state of program configuration.
- The data moves freely in a program.
- Procedural programs model real-world processes as 'procedures' operating on 'data'.
- It is easy to follow the logic of a program.

Languages: C, Assembly etc.

```python
# This is a simple program to find factorial of the 
n = int(input("Enter a number: "))
product = 1
while n > 0:
    product *= n
    n -= 1
print(product)
```

# Object-Oriented Programming

#### In object oriented programming, program is divided into small parts called objects.
Objects contain data in the form of attributes and code in the form of methods. 

### Characterstics
- Data Abstraction
- Data Encapsulation 
- Inheritance 
- Polymorphism

Languages: C++, Simula, Java etc.

```python
class FactorialCalculator:
    def __init__(self, n: int):
        self.n: int = n

    def calculate(self) -> int:
        product: int = 1
        while self.n > 0:
            product *= self.n
            self.n -= 1
        return product
```

```python
F1 = FactorialCalculator(5)
F1.calculate()
```

# Functional Programming

 #### Functional programming is a programming paradigm where programs are constructed by applying and composing functions. 
 This style of programming is the most close bridge to the mathematical definitions. 
 
 ### Characterstics
- Variables are just the means of abstraction.
- The functions are first class components and pure(without any side effect).
- Functional programming languages donâ€™t support flow Controls like loop statements and conditional statements like If-Else and Switch Statements. They directly use the functions and functional calls.

Languages: Haskell, Scheme(LISP) 

```python
def fact(n: int) -> int:
    if n == 0 or n == 1:
        return 1
    return n * fact(n - 1)
    
```

```python
fact(6)
```
